## 🔹 5. Data Cleaning

Before analysis, we need to clean the dataset to ensure all values are in the correct format.  

### 🧹 Cleaning Steps:
1. **Currency Columns**:  
   - Columns such as `Cost Price`, `Retail Price`, `Profit Margin`, `Sub Total`, `Discount $`, `Order Total`, `Shipping Cost`, and `Total` contain a `$` sign and are stored as text (object type).  
   - We will **remove the `$` sign** and convert these columns into **numeric (float)** values.  

2. **Percentage Column**:  
   - The `Discount %` column contains a `%` sign and is stored as text.  
   - We will **remove the `%` sign** and convert it into a numeric value.  

3. **Missing Values**:  
   - There are only 2 missing values (`Address` and `Order Quantity`).  
   - We will handle them later during the cleaning/analysis step.  

After cleaning, our dataset will be ready for **statistical analysis and visualization**.


```python
# Columns that contain $ sign (money values)
money_cols = [
    "Cost Price", "Retail Price", "Profit Margin",
    "Sub Total", "Discount $", "Order Total",
    "Shipping Cost", "Total"
]

# Remove $ and convert to float
for col in money_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)

# Clean percentage column (remove % and convert to float)
df["Discount %"] = df["Discount %"].replace('%', '', regex=True).astype(float)

# Preview cleaned dataset
df[money_cols + ["Discount %"]].head()
```

## 🔹 8. RFM (Recency, Frequency, Monetary) Analysis  

RFM analysis is a marketing technique to evaluate customer value by looking at three key factors:  

1. **Recency (R):** How recently a customer made a purchase.  
2. **Frequency (F):** How often a customer makes a purchase.  
3. **Monetary (M):** How much money a customer spends.  

👉 This helps businesses identify **loyal customers**, **big spenders**, and those who may be at risk of churn.  


```python
import pandas as pd

# 🔹 Ensure dates are in datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# 🔹 Set a reference date (max order date in dataset)
reference_date = df["Order Date"].max()

# 🔹 Calculate RFM values
rfm = df.groupby("Customer Name").agg({
    "Order Date": lambda x: (reference_date - x.max()).days,  # Recency
    "Order No": "count",                                     # Frequency
    "Total": "sum"                                           # Monetary
})

# 🔹 Rename columns
rfm.rename(columns={
    "Order Date": "Recency",
    "Order No": "Frequency",
    "Total": "Monetary"
}, inplace=True)

# 🔹 Preview RFM table
print(rfm.head())
```

## 🔹 9. RFM Scoring & Customer Segmentation  

After calculating **Recency, Frequency, and Monetary**, we assign scores (1–5) to each metric:  

- **Recency (R):** Lower is better (recent buyers).  
- **Frequency (F):** Higher is better (frequent buyers).  
- **Monetary (M):** Higher is better (big spenders).  

Then we combine R, F, and M scores into a single **RFM Score** and segment customers into groups such as:  

- **Champions** – Recent, frequent, and high spenders.  
- **Loyal Customers** – Buy often but not always high spend.  
- **At Risk** – Haven’t purchased recently.  
- **Hibernating** – Old, inactive customers.  
- **Big Spenders** – Not frequent but high spend.  


```python
import pandas as pd

# 🔹 Step 1: Calculate quantiles for RFM segmentation
quantiles = rfm.quantile(q=[0.2, 0.4, 0.6, 0.8])
quantiles

# 🔹 Step 2: Functions to assign scores
def r_score(x):
    if x <= quantiles['Recency'][0.2]:
        return 5
    elif x <= quantiles['Recency'][0.4]:
        return 4
    elif x <= quantiles['Recency'][0.6]:
        return 3
    elif x <= quantiles['Recency'][0.8]:
        return 2
    else:
        return 1

def fm_score(x, col):
    if x <= quantiles[col][0.2]:
        return 1
    elif x <= quantiles[col][0.4]:
        return 2
    elif x <= quantiles[col][0.6]:
        return 3
    elif x <= quantiles[col][0.8]:
        return 4
    else:
        return 5

# 🔹 Step 3: Apply scores
rfm['R'] = rfm['Recency'].apply(r_score)
rfm['F'] = rfm['Frequency'].apply(lambda x: fm_score(x, 'Frequency'))
rfm['M'] = rfm['Monetary'].apply(lambda x: fm_score(x, 'Monetary'))

# 🔹 Step 4: Create RFM Score
rfm['RFM_Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)

# 🔹 Step 5: Define segmentation rules
def segment_customer(df):
    if df['R'] >= 4 and df['F'] >= 4 and df['M'] >= 4:
        return 'Champion'
    elif df['R'] >= 3 and df['F'] >= 3:
        return 'Loyal Customer'
    elif df['R'] >= 4 and df['F'] <= 2:
        return 'Big Spender'
    elif df['R'] <= 2 and df['F'] >= 3:
        return 'At Risk'
    else:
        return 'Hibernating'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)

# 🔹 Step 6: Preview results
print(rfm.head())
print(rfm['Segment'].value_counts())
```

### 8. Average Profit Margin by Order Size

To better understand how order size affects profitability, we grouped orders into three categories:
- **Small Orders**: 1–15 items  
- **Medium Orders**: 16–30 items  
- **Large Orders**: 31+ items  

We then calculated the **average profit margin** for each category and plotted it as a bar chart.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your complete dataset
df = pd.read_csv("data.csv")

# Clean Profit Margin column (remove $ and convert to float)
df["Profit Margin"] = df["Profit Margin"].replace('[\$,]', '', regex=True).astype(float)

# Ensure Order Quantity is numeric
df["Order Quantity"] = pd.to_numeric(df["Order Quantity"])

# Step 1: Define order size categories
def order_size(qty):
    if qty <= 15:
        return "Small"
    elif qty <= 30:
        return "Medium"
    else:
        return "Large"

# Step 2: Apply categorization
df['Order Size'] = df['Order Quantity'].apply(order_size)

# Step 3: Calculate average profit margin by order size
avg_profit_margin = df.groupby('Order Size')['Profit Margin'].mean().reset_index()

# Step 4: Bar chart visualization
plt.figure(figsize=(8,6))
plt.bar(avg_profit_margin['Order Size'], avg_profit_margin['Profit Margin'], 
        color=['#4CAF50','#2196F3','#FF9800'])
plt.xlabel("Order Size Category")
plt.ylabel("Average Profit Margin")
plt.title("Average Profit Margin by Order Size")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Insights: Average Profit Margin by Order Size

- From the bar chart, we observe that **Large orders have the highest average profit margin**, followed by **Small orders**, and **Medium orders have the lowest**.
- This suggests that profitability is not directly proportional to order quantity; instead, there may be **pricing strategies or product selection differences** influencing profit margins.

### Possible Reasons

1. **Large orders**:
   - Likely include high-value or high-margin products.
   - Economies of scale may reduce shipping or packaging costs, increasing profit margin.
2. **Small orders**:
   - Could be frequent repeat purchases of high-margin items, or premium-priced products bought in small quantities.
3. **Medium orders**:
   - May include bulk purchases of lower-margin products, or discounted items, reducing the average profit margin.

### Recommendations

- **Focus marketing and promotions on high-margin large orders** to maximize overall revenue.
- **Analyze Medium orders** to identify why their profit margins are lower; consider adjusting pricing or targeting more profitable products in this category.
- **Optimize product mix** by promoting small high-margin products alongside large orders to maintain profitability.
- Monitor order size trends over time to adapt inventory and pricing strategies accordingly.


## 9: Top 5 Most Profitable Products
- Sorted products by `Total Profit` in descending order.
- Displayed the top 5 products contributing the most to revenue.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your complete dataset
df = pd.read_csv("data.csv")

# Clean monetary columns (remove $ and convert to float)
monetary_cols = ["Profit Margin"]
for col in monetary_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)

# Ensure Order Quantity is numeric
df["Order Quantity"] = pd.to_numeric(df["Order Quantity"])

# Step 1: Calculate Total Profit per row
df["Total Profit"] = df["Profit Margin"] * df["Order Quantity"]

# Step 2: Aggregate total profit by Product Name
product_profit = df.groupby("Product Name")["Total Profit"].sum().reset_index()

# Step 3: Find the top 5 most profitable products
top5_products = product_profit.sort_values(by="Total Profit", ascending=False).head(5)
print("Top 5 Most Profitable Products:")
print(top5_products)

# Step 4: Find the product with maximum total profit
max_profit_product = product_profit.loc[product_profit["Total Profit"].idxmax()]
print("\nProduct with Maximum Total Profit:")
print(f"Product Name: {max_profit_product['Product Name']}")
print(f"Total Profit: ${max_profit_product['Total Profit']:,.2f}")

# Step 5: Visualize Top 5 Products
plt.figure(figsize=(10,6))
bars = plt.bar(top5_products["Product Name"], top5_products["Total Profit"], color="skyblue")
bars[0].set_color("orange")  # Highlight the most profitable product

plt.title("Top 5 Most Profitable Products")
plt.ylabel("Total Profit ($)")
plt.xticks(rotation=30, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Add value labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10, f"${yval:,.0f}", ha="center", va="bottom")

plt.show()
```

## Insights and Recommendations
- The highlighted product contributes the most to overall revenue and should be prioritized in inventory and marketing decisions.
- Consider analyzing high-profit products by customer type and region for targeted sales strategies.
- Further analysis can include discounts, order quantities, and seasonal trends to optimize revenue.


## Total Profit by Shipping Mode

This analysis explores how different shipping modes contribute to the overall profit. 

- The `Ship Mode` column indicates the shipping method chosen for each order.
- `Total Profit` is calculated per order as `Profit Margin × Order Quantity`.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv("data.csv")

# Clean Profit Margin column
df["Profit Margin"] = df["Profit Margin"].replace('[\$,]', '', regex=True).astype(float)

# Ensure numeric
df["Order Quantity"] = pd.to_numeric(df["Order Quantity"])
df["Shipping Cost"] = df["Shipping Cost"].replace('[\$,]', '', regex=True).astype(float)

# Calculate Total Profit per row
df["Total Profit"] = df["Profit Margin"] * df["Order Quantity"]

# Group by shipping mode
shipping_summary = df.groupby("Ship Mode")["Total Profit"].sum().reset_index().sort_values(by="Total Profit", ascending=False)

# Plot total profit by shipping mode
plt.figure(figsize=(8,6))
plt.bar(shipping_summary["Ship Mode"], shipping_summary["Total Profit"], color=['#2196F3', '#4CAF50', '#FF9800'])
plt.xlabel("Shipping Mode")
plt.ylabel("Total Profit ($)")
plt.title("Total Profit by Shipping Mode")
plt.xticks(rotation=30, ha="right")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Insights: Total Profit by Shipping Mode

After analyzing the dataset:

- **Regular Air** contributes the **highest total profit**.
- **Express Air** comes second in total profit.
- **Delivery Truck** contributes the **least total profit**.

### Possible Reasons

1. **Regular Air**: Likely used for a large volume of orders, including both small and medium-sized products, which collectively generate high total profit.  
2. **Express Air**: Faster shipping may be chosen for high-margin or urgent products, but fewer orders compared to Regular Air result in slightly lower total profit.  
3. **Delivery Truck**: Typically used for local or bulk shipments of lower-margin products, resulting in the lowest total profit.

### Recommendations

- **Focus on Regular Air orders**: Maintain efficient handling and customer service to continue maximizing profits.  
- **Promote high-margin products via Express Air**: Encourage customers to use faster shipping for premium products to boost profit.  
- **Review Delivery Truck usage**: Consider optimizing routes or promoting higher-margin products for this mode to increase profitability.  
- **Balance shipping costs and profit margins**: Monitor shipping mode trends to make data-driven logistics and pricing decisions.
